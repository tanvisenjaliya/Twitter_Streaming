### Now Taking tweets which are in English Language

```python
df2=[]
for tweet in range(len(df1)):
    if df1.lang[tweet]=="en":
        df2.append(df1.loc[tweet,])
df2 = pd.DataFrame(df2) 
df2
```

# Out of 20,000 we got 7284 tweets

### Cleaning of tweets by removing special characters

```python
def clean_text(df, text_field):
    #df[text_field] = df[text_field].str.lower()
    df[text_field] = df[text_field].apply(lambda elem: re.sub("(RT @\w+:)|(RT+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", elem)) 
    return df
df2 = clean_text(df2, 'text')

```
