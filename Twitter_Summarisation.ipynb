# Summarize
The parameters are:

ratio: It can take values between 0 to 1. It represents the proportion of the summary compared to the original text.
word_count: It decides the no of words in the summary.

```python
summary_by_ratio=summarize(original_text,ratio=0.5)
print(summary_by_ratio)
```

```python
# Summarization by word count
summary_by_word_count=summarize(original_text,word_count=30)
print(summary_by_word_count)
```

```python
# Summarization when both ratio & word count is given
summary=summarize(original_text, ratio=0.5, word_count=30)
print(summary)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
# Import the LexRank summarizer
from sumy.summarizers.lex_rank import LexRankSummarizer
```

```python
# Initializing the parser
my_parser = PlaintextParser.from_string(original_text,Tokenizer('english'))

# Creating a summary of 3 sentences.
lex_rank_summarizer = LexRankSummarizer()
lexrank_summary = lex_rank_summarizer(my_parser.document,sentences_count=3)

# Printing the summary
for sentence in lexrank_summary:
  print(sentence)
```

```python
# Import the summarizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python
# creating the summarizer
lsa_summarizer=LsaSummarizer()
lsa_summary= lsa_summarizer(my_parser.document,5)

# Printing the summary
for sentence in lsa_summary:
    print(sentence)
```

```python
# Import the summarizer
from sumy.summarizers.luhn import LuhnSummarizer
```

```python
#  Creating the summarizer
luhn_summarizer=LuhnSummarizer()
luhn_summary=luhn_summarizer(my_parser.document,sentences_count=3)

# Printing the summary
for sentence in luhn_summary:
  print(sentence)
```

```python
from sumy.summarizers.kl import KLSummarizer
```

```python
# Instantiating the  KLSummarizer
kl_summarizer=KLSummarizer()
kl_summary=kl_summarizer(my_parser.document,sentences_count=3)

# Printing the summary
for sentence in kl_summary:
    print(sentence)
```

```python
import os
import numpy as np
from os import path
from wordcloud import ImageColorGenerator
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image

from wordcloud import STOPWORDS
```

```python
text = open('AppleEssay.txt',encoding="utf8").read()

#now not needed
mask = np.array(Image.open("AppleLogo1.jpg"))
wordcloud_usa = WordCloud(stopwords=STOPWORDS, background_color="black",min_font_size=2,
                          width= 1600, height= 800, 
                          mode='RGB', max_words=4000, mask=mask).generate(text)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[12,12])
plt.imshow(wordcloud_usa.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

# store to file
plt.savefig("Apple.jpg", format="jpg")

plt.show()
```
