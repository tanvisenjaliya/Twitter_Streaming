## Fetching the requied data from json file and storing it into csv file.

```python
import pandas as pd
import json
import re
```

```python
tweets_data_path='AppleEvent1.txt'
tweets_data=[]
tweets_file=open(tweets_data_path,"r")

for line in tweets_file:
    tweet=json.loads(line)
    tweets_data.append(tweet)
```

```python
df1 = pd.DataFrame(tweets_data, columns=['created_at','text','user','lang'])

df1['name'] = [d['name']for d in df1.user]
df1['friends_count'] = [d['friends_count']for d in df1.user]
df1['screen_name'] = [d['screen_name']for d in df1.user]
df1['location'] = [d['location']for d in df1.user]
df1['followers_count'] = [d['followers_count']for d in df1.user]
df1['favourites_count'] = [d['favourites_count']for d in df1.user]
df1['listed_count'] = [d['listed_count']for d in df1.user]
df1['statuses_count'] = [d['statuses_count']for d in df1.user]
```
