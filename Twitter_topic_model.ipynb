```python
# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
#from gensim.models import CoherenceModel
```

```python
# Plotting tools
import pyLDAvis

import matplotlib.pyplot as plt
%matplotlib inline

import pyLDAvis.gensim_models

import pandas as pd
from nltk import word_tokenize
```

```python
data=pd.read_csv('cleantweets.csv')
```

```python
data
```

```python
def remove_punctuation(texts):
    punc = '''!()-[]{};:'"\\,<>./?@#$%^&*_~'''
    
    for ele in texts:
        if ele in punc:
            texts = str(texts).replace(ele, "")
    return texts
```

```python
data_pt=[]
for i in range(len(data.text)):
    data_pt.append(remove_punctuation(data.text[i]))
```

```python
data.text
```

# Tokenize

```python
data_tk=[]
for i in range(len(data)):
    data_tk.append(word_tokenize(data.text[i]))
```

```python
data_tk
```

# Removing Stopwords

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```

```python
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]
```

```python
data_sw=[]
for i in range(len(data_tk)):
    data_sw.append(remove_stopwords(data_tk[i]))
```

```python
data_sw
```

```python
data_doc = []
for outer_item in data_sw:
    inner_list = []
    for inner_item in outer_item:
        for innermost_item in inner_item:
            inner_list.append(innermost_item)
    data_doc.append(inner_list)
    
```

```python
print(data_doc)
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_doc)

# Create Corpus
texts = data_doc

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1])
```

Gensim creates a unique id for each word in the tweets text document. The produced corpus is showing a mapping of (word_id, word_frequency).

For example, (0, 1) above implies, word id 0 occurs once in the first document. Likewise, word id 2 occurs once and so on.

```python
id2word[1]
```

```python
# Human readable format of corpus (term-frequency)
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

```python
# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=20, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=500,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# Print the Keyword in the 10 topics
from pprint import pprint
pprint(lda_model.print_topics())
```

```python
doc_lda = lda_model[corpus]
```

```python
doc_lda
```

```python
pprint(doc_lda)
```

```python
from gensim.models import CoherenceModel
```

```python
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  # It is a measure of how good the model is. The lower value means  the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=data_doc, dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
